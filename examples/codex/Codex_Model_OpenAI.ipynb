# The main Program using Codex model of openAI.

```python
import os
import openai

Query = input('Ask me a question: ')
print('Mostafizur')

openai.api_key = ("Give your API-key here.")

response = openai.Completion.create(
  model="code-davinci-002",
  prompt="",
  temperature=0,
  max_tokens=256,
  top_p=1,
  frequency_penalty=0,
  presence_penalty=0
)
content = response.choices[0].text
print(content)
Query = content
```
